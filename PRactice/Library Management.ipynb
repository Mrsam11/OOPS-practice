```python
class Employee:
    def __init__(self,first_name,last_name,salary):
        self.first_name = first_name
        self.last_name = last_name
        self.salary = salary
        
    def display(self):
        print(f"First name is {self.first_name} \nLast name is {self.last_name} and salary is {self.salary}")
        
    def increment(self,change):
        self.salary = change
```

```python
emp1 = Employee("Sameer","Ahmed",10000) #Emp1 is an object of the class not variable
emp1.display() #Calling class
emp1.increment(20000)
emp1.display()
print(emp1.salary) #for printing single item from class 
```

```python
class Library:
    def __init__(self, listofbooks):
        self.available = listofbooks
        
    def display(self):
        print("Available books in library are : ")
        for book in self.available:
            print(book)
            
    def lend(self, requested):
        if requested in self.available:
            print(f"Now you can have the book {requested}")
            self.available.remove(requested)
        else:
            print(f"The requested book {requested} is not available in library")
            
    def add_book(self, adding):
        if adding in self.available:
            print("We Already have that book")
        else:
            print("Thank you for book")
            self.available.append(adding)
    
class Customer:
    def requesting(self):
        self.book = input("Enter The Name Of the Book that you Wanted")
        return self.book
    def returning(self):
        self.book = input("Enter name of the book that you want to return")
        
    
library = Library(["Harry Potter and sorceror stone", "Six of crows", "the witcher","Goblin planet","a smarter way to learn"])
customer = Customer()

while True:
    print("Enter 1 to display book")
    print("Enter 2 to lend")
    print("Enter 3 to return")
    print("enter 4 to quit")
    
    userchoice= int(input())
    
    if userchoice == 1:
        library.display()
    elif userchoice == 2:
        requested_book = customer.requesting()
        library.lend(requested_book)
    elif userchoice == 3 :
        returned = customer.returning()
        library.add_book(returned)
    elif userchoice == 4:
        quit()
```
